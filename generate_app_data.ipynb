```python
import pandas as pd

pitches_df = pd.read_csv("all_pitches.csv")

def process_batters(pitches_df, batters):
    batter_mapping = {}
    
    for batter_group in batters:
        primary_batter_id = float(batter_group[0])
        for batter_id in batter_group[1:]:
            batter_mapping[float(batter_id)] = primary_batter_id
    
    pitches_df['BatterId'] = pitches_df['BatterId'].astype(float)
    pitches_df.loc[pitches_df['BatterId'].isin(batter_mapping), 'BatterId'] = pitches_df['BatterId'].map(batter_mapping).fillna(pitches_df['BatterId'])

    primary_batter_ids = {float(group[0]) for group in batters}
    filtered_batter_df = pitches_df[pitches_df['BatterId'].isin(primary_batter_ids)]

    return filtered_batter_df


def process_pitchers(pitches_df, pitchers):
    pitcher_mapping = {}
    
    for pitcher_group in pitchers:
        primary_pitcher_id = float(pitcher_group[0])
        for pitcher_id in pitcher_group[1:]:
            pitcher_mapping[float(pitcher_id)] = primary_pitcher_id
    
    pitches_df['PitcherId'] = pitches_df['PitcherId'].astype(float)
    pitches_df.loc[pitches_df['PitcherId'].isin(pitcher_mapping), 'PitcherId'] = pitches_df['PitcherId'].map(pitcher_mapping).fillna(pitches_df['PitcherId'])

    primary_pitcher_ids = {float(group[0]) for group in pitchers}
    filtered_pitches_df = pitches_df[pitches_df['PitcherId'].isin(primary_pitcher_ids)]

    return filtered_pitcher_df

### DO NOT TOUCH THE CODE ABOVE
### PUT PITCHER IDS AND BATTER IDS HERE
pitchers = [
            [823227, 1000121636], #Zeldin            
            [809709, 1000051331], ### Charlie Goldstein
            [1000110601], ### Alton Davis
            [815145, 10108696], ### Bradley Stewart
            [809714, 1000096939], ### Collin Caldwell
            [90000235966], ### Zach Harris
            [1000295488], ### Jordan Stephens
            [10056544, 1000175820], ### Brian Curley
            [1000241035, 1000164129, 10053243], ### Davis Chastain
            [803287, 1000110595], ### Leighton Finley
            [1000113633], ### Asher Sabom
            [1000088474, 702614], ### Matthew Hoskins
            [1000187635, 701368], ### Kolten Smith
            [1000233770, 10006526], ### JT Quinn
            [1000092911, 695727], ### Eric Hammond
            [90000256197], ### Logan Spivey
            [10268112], ### Nate Taylor
            [823230, 1000076701], ### Tyler McLoughlin
            [1000269766, 10086131], ### Paul Farley
            [10276837, 1000057096, 1000334133], ### Zach Brown
            [809712, 1000066681], ### DJ Radtke
            [812810, 10064790], ### Wyatt Land
            [1000208929], ### Justin Byrd
            [823232, 10053126], ### Luke Wiltrakis
            [10104083], ### Lucas Morici
            [90000217140] ### Ethan Sutton
]

### PUT BATTER IDS HERE
batters = [
            #[1000069469, 695773,1000391664,1000390931], ### Vidal Colon
            [1000075143], ### Jackson Bessette
            #[695477, 1000052101], ### Blake Cowans
            #[803325, 1000051217], ### Brady Patterson
            [1000235203], ### Jack Snyder
            [1000115808], ### Thomas Curry
            #[10000375, 1000236751], ### Colin Husko
            #[1000255777], ### Christian Holmes
            [10069733], ### AJ Taylor
            #[823754, 1000119891], ### Alex Mezzetti
            [10008685, 1000235233], ### DJ Butler
            [1000099592], ### James Harris
            #[1000255611], ### Will Flanigan
            [1000268592], ### Pambos Nicoloudes
            #[1000075636], ### Jonathan Fleaka
            #[90000255731, 828714], ### Jake Troyner
            #[274214, 1000232995], ### David Cox?
            [1000114430], ### Terrick Thompson-Allen
            #[1000209296], ### Lucas Smith
            [1000074821, 804114] ### Luke Nowak
]


filtered_batter_df = process_batters(pitches_df, batters)
filtered_pitcher_df = process_batters(pitches_df, pitchers)

combined_df = pd.concat([filtered_batter_df, filtered_pitcher_df]).drop_duplicates()
combined_df = combined_df.reset_index(drop=True)

#New comment2

combined_df.to_csv('app_data.csv')
```
